# Технологический процесс

**Cтадии процесса.**

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

<img src="https://pictures.s3.yandex.net/resources/viruchka_1576238830.jpg" alt="drawing" width="900"/>

**1. Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

#### Создание функций по обработке данных

**Функция по выводу дубликатов.** 

* на вход подается датафрейм,
* выходные данные - количество дубликатов.

**Функция по выводу пропущенных значений.** 

* на вход подается датафрейм,
* выходные данные - пропущенные значения.

*P.S. Цикл внутри функции позволяет срезать столбцы, где пропущенные значения равны нулю.*

**Функция для заполнения пропусков.** 

* на вход подается датафрейм,
* выходные данные - отсутствуют, но столбцы с пропущенными значениями заполнены.

*P.S. Первый цикл идет до последнего столбца, где есть пропуски и загружает в список названия столбцов.*

*Второй цикл проходит по списку и подает столбцы в функцию для их заполнения **медианой**.*

```python
#Поиск дубликтов
def duplicated_data(data):   
    duplic = data.duplicated().sum()
    return duplic 

# Пропущенные значения 
def pass_data(data):
    count = 0
    data_isna = data.isna().sum().sort_values(ascending=False)
    for _ in range(len(data_isna)):
        if data_isna[_] == 0:
            count += 1
            
    data_isna = data_isna[:-count]
    return data_isna

#Заполнение пропусков    
def fill(data, columns):
    data[columns] = (data[columns].fillna(data[columns].median()))
    
def elimination(data):
    list_ = []
    for _ in range(pass_data(data).shape[0]):
        pass_data(data).index[_]
        list_.append(pass_data(data).index[_])

    for _ in range(len(list_)):
        fill(data, list_[_])
```

## Модель

### sMAPE

Напишим функцию для вычисления итоговой sMAPE.

<img src="https://pictures.s3.yandex.net/resources/smape_1576239058.jpg" alt="drawing" width="800"/>

<img src="https://pictures.s3.yandex.net/resources/_smape_1576239054.jpg" alt="drawing" width="800"/>

```python
def smape(y_target, y_pred):
    n = len(y_target)
    smape = np.sum(np.abs(y_target - y_pred) / ((np.abs(y_target) + np.abs(y_pred)) / 2)) * 100 / n    

    return smape
```

```python
def final_smape(rougher, final):
    final_smape = round((.25 * rougher + .75 * final), 2)
    return final_smape
```

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
def calc_smape(model, X, y):
    score = cross_val_score(model, X, y, scoring=smape_scorer, cv=5) 
    final_score = round(abs(score.mean()), 2)
    
    return final_score
```

```python
def info_model(rougher, final):
    print(f'sMAPE черновой концетрат: \033[1m{rougher}\033[0m')
    print(f'sMAPE финальный концетрат: \033[1m{final}\033[0m')
    print(f'Финальный sMAPE: \033[1m{final_smape(rougher, final)}\033[0m')
```

```python
def best_score_model(model, rougher):
    if model == '<class sklearn.dummy.DummyRegressor>':
        best_result = 0
        best_model = None
        if rougher > best_result:
            best_result = rougher
            best_model = model
    else:         
        return rougher
```

**Дерево решений.**

```python
%%time 

best_model_tree_r = None
best_result_tree_r = 0 


best_model_tree_f = None
best_result_tree_f = 0 

for depth in range(1, 5):
    model_tree = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    
    scores_tree_r = cross_val_score(model_tree, X_train_rougher, y_train_rougher, scoring=smape_scorer, cv=5) 
    final_score_tree_r = abs(scores_tree_r.mean())
    
    if final_score_tree_r > best_result_tree_r:
            best_result_tree_r = round(final_score_tree_r, 2)
            best_model_tree_r = model_tree
            
            
    scores_tree_f = cross_val_score(model_tree, X_train_final, y_train_final, scoring=smape_scorer, cv=5) 
    final_score_tree_f = abs(scores_tree_f.mean())
    
            
    if final_score_tree_f > best_result_tree_f:
            best_result_tree_f = round(final_score_tree_f, 2)
            best_model_tree_f = model_tree

print(f'Лучшая модель:{best_model_tree_r}')
print(f'sMAPE дерева решений, черновой концетрат: {best_result_tree_r}')
print()
print(f'Лучшая модель:{best_model_tree_f}')
print(f'sMAPE дерева решений, финальный концетрат: {best_result_tree_f}')
print()
print('Финальный sMAPE:', final_smape(best_result_tree_r, best_result_tree_f))
print()
```

```python
%%time 

model_regres = LinearRegression()

scores_regres_r = cross_val_score(model_regres, X_train_rougher, y_train_rougher, scoring=smape_scorer, cv=5) 
final_score_regres_r =  round(abs(scores_regres_r.mean()), 2)

print('sMAPE линейной регрессии, черновой концетрат:', final_score_regres_r)
print()

scores_regres_f = cross_val_score(model_regres, X_train_final, y_train_final, scoring=smape_scorer, cv=5) 
final_score_regres_f =  round(abs(scores_regres_f.mean()), 2)

print('sMAPE линейной регрессии, финальный концетрат:', final_score_regres_f)
print()
print('Финальный sMAPE ', final_smape(final_score_regres_r, final_score_regres_f))
```

**Модель случайного леса.**

```python
%%time 

best_model_forest_r = None
best_result_forest_r = 0 


best_model_forest_f = None
best_result_forest_f = 0 

for n_est in range(1, 6, 2):
    for depth in range(1, 5):
        model_forest = RandomForestRegressor(random_state=12345, n_estimators=n_est, max_depth=depth)

        scores_forest_r = cross_val_score(model_forest, X_train_rougher, y_train_rougher, scoring=smape_scorer, cv=5) 
        final_score_forest_r = abs(scores_forest_r.mean())

        if final_score_forest_r > best_result_forest_r:
                best_result_forest_r = round(final_score_forest_r, 2)
                best_model_forest_r = model_forest


        scores_forest_f = cross_val_score(model_forest, X_train_final, y_train_final, scoring=smape_scorer, cv=5) 
        final_score_forest_f = abs(scores_forest_f.mean())


        if final_score_forest_f > best_result_forest_f:
                best_result_forest_f =  round(final_score_forest_f, 2)
                best_model_forest_f = model_forest

print(f'Лучшая модель:{best_model_forest_r}')
print(f'sMAPE дерева решений, черновой концетрат: {best_result_forest_r}')
print()
print(f'Лучшая модель:{best_model_forest_f}')
print('sMAPE дерева решений, финальный концетрат:', best_result_forest_f)
print()
print('Финальный sMAPE ', final_smape(best_result_forest_r, best_result_forest_f))
```
