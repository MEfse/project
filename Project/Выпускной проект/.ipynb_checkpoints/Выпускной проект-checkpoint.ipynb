# Описание проекта 

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Вам предстоит построить модель, которая предскажет температуру стали.

# Описание этапа обработки

Сталь обрабатывают в металлическом ковше **вместимостью около 100 тонн**. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз **ввести легирующие добавки, измеряют температуру стали и производят её химический анализ.** Потом **температуру** на несколько минут **повышают**, добавляют легирующие материалы и продувают сплав инертным газом. Затем его **перемешивают** и снова проводят измерения. Такой цикл **повторяется до достижения целевого химического состава и оптимальной температуры плавки.**

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

## Изменения типов данных 
<a name="типы_данных"></a>

[Возврат к плану действий](#план)

Необходимо привести тип данных к временному. 

```python
def change_data_type(data, *args):
    ''' 
    Casting data types to a temporary 
    
    Args:
        data (pandas.core.frame.DataFrame): Dataframe with wrong data type
        
    Returns:
        data (pandas.core.frame.DataFrame): Dataframe with correct data type
    '''
    data = data.copy()
    columns = data.columns[1:]
    
    #Если в args переданы столбцы, только они подлежат изменению
    if len(args) != 0:
        for _ in range(len(args)):
            data[args[_]] = pd.to_datetime(data[args[_]], format='%Y-%m-%d %H:%M:%S')
        
        return data
    
    #Если только датафрейм, то все столбцы кроме первого(key) будут изменены
    else:
        for _ in columns:
            data[_] = pd.to_datetime(data[_], format='%Y-%m-%d %H:%M:%S')
        
        return data
```

```python
def build_graph(data, x, y, axhline=False):   
    ''' 
    Plotting with a specific set of parameters 
    
    Args:
        data (pandas.core.frame.DataFrame): Dataframe for plotting
        x (str): Dataframe column to display on the x-axis
        y (str): Dataframe column to display on the y-axis
        axhline (bool): Boolean value for plotting the average on the graph
    '''
    
    data.plot(x, 
              y, 
              grid=True, 
              figsize=(12, 8), 
              kind='scatter');
    
    if axhline == True: 
        avg = round(data[y].mean(), 3)
        print('\033[1mСреднее значение:\033[0m', avg)
        plt.axhline (y=np.nanmean(avg), color='red');
```

**График в нуле сильно прижался из-за аномалий** в данных, от которых нужно избавиться. **Удалять аномалии следует по ключу**, для того чтобы часть аномальных данных так же удалилась. 

**Найдем ключ аномалии.**

```python
def remove_anomaly(data, columns='unknown', *value):
    ''' 
    Removing anomalies in data 
    
    Args:
        data (pandas.core.frame.DataFrame): Dataframe to remove anomalies
        columns (str): The column in which the anomaly is located
        value (int): Threshold at which values will be considered an anomaly
        
    Returns:
        data (pandas.core.frame.DataFrame): Dataframe cleared of anomalies
    '''
        
    data = data.copy()
    
    #В случае если порог для аномалий известен 
    if len(value) < 0:
        data_anomaly = data.loc[data[columns] < value]
        index_anomaly = data_anomaly['key'].to_list()
    
        data = delete_index(data, index_anomaly)
            
        return data
    
    #В случае если порог для аномалий заранее нельзя просчитать (для графиков)
    else:
        #Для одного признака
        if columns != 'unknown':
            index_anomaly = get_interquarter_range(data, columns)
        
            data = delete_index(data, index_anomaly)
            
            return data
        else:
            #Для всех признаков (исключение key)
            columns = data.columns[1:]
            for _ in columns:
                index_anomaly = get_interquarter_range(data, _)
        
                data = delete_index(data, index_anomaly)
            
            return data
```

```python
def get_interquarter_range(data, columns):
    ''' 
    Сallback function - remove_anomaly
    
    Args:
        data (pandas.core.frame.DataFrame): Dataframe to remove anomalies
        columns (str): The column in which the anomaly is located
        
    Returns:
        key (list): Anomalous keys
    '''
    Q1 = data[columns].quantile(0.25)
    Q3 = data[columns].quantile(0.75)
    IQR = Q3 - Q1
    key = data[(data[columns] < Q1-1.5*IQR ) | (data[columns] > Q3+1.5*IQR)]['key'].to_list()
    
    return key
```

```python
def delete_index(data, key):
    ''' 
    Сallback function - remove_anomaly
    
    Args:
        data (pandas.core.frame.DataFrame): Dataframe to remove anomalies
        columns (str): The column in which the anomaly is located
        
    Returns:
        data (pandas.core.frame.DataFrame): Dataframe cleared of anomalies
    '''
       
    for k in key:
        data = data.loc[data['key'] != k]
        
    return data
```

### Вывод

Построим общую таблицу по метрике МАЕ

```python
mae_models = pd.DataFrame({'Дамми-регрессия' : [mae_dummy, ],
                            'Случайный лес' : [abs(grid_forest.best_score_)],
                            'Линейная регрессия' : [abs(grid_linear.best_score_)],
                            'Кэтбуст' : [abs(grid_cat.best_score_)]}).T

mae_models.columns = ['MAE']
mae_models = mae_models.sort_values('MAE', ascending=True)
```

```python
display(mae_models)
```

**Тест наилучшей модели**

```python
def test_model(name_model, model, test):
    ''' 
    testing models 
    
    Args:
        name_model (list): Name of the best model
        model (class 'sklearn.model_selection._search.GridSearchCV'): The model to be tested
        test (class 'pandas.core.frame.DataFrame'): Test data
    '''
    
    try:
        predict = model.predict(test)
        print(name_model)
        print("Используемая метрика", model.scorer_)
        print("Лучшие параметры", model.best_params_)
        print("МАЕ тестовых данных", mean_absolute_error(predict, y_test).round(2))
        
    except:
        predict = model.predict(test)
        print("МАЕ тестовых данных", mean_absolute_error(predict, y_test).round(2))
```

```python
best_model = mae_models.head(1).index.values
```

```python
if best_model == 'Кэтбуст': 
    test_model(best_model, grid_cat, X_test)
    
elif best_model == 'Случайный лес':
    test_model(best_model, grid_forest, X_test)
    
elif best_model == 'Линейная регрессия':
    test_model(best_model, grid_linear, X_test_scaled)
    
else:
    test_model(best_model, model_dummy, X_test)
```
